```python
import matplotlib.pyplot as plt

# Assuming your text data is in 'xyz_features.txt' and 'hks_features.txt'

def read_data(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()[1:]  # Skip the header line
        data = [list(map(float, line.strip().split(';'))) for line in lines]
    return data

# Read the data from files
xyz_data = read_data('diffnet_xyz.txt')
hks_data = read_data('diffnet_hks.txt')
shot_data = read_data('diffnet_shot.txt')
lps_data = read_data('diffnet_lps.txt')
mma_data = read_data('diffnet_mma.txt')


# Convert the data to a format suitable for plotting
# Transposing the list of lists to get columns of data
xyz_data_transposed = list(map(list, zip(*xyz_data)))
hks_data_transposed = list(map(list, zip(*hks_data)))
shot_data_transposed = list(map(list, zip(*shot_data)))
lps_data_transposed = list(map(list, zip(*lps_data)))
mma_data_transposed = list(map(list, zip(*mma_data)))

print('xyz',xyz_data_transposed)
print('hks',hks_data_transposed)
print('shot',shot_data_transposed)
print('lps',lps_data_transposed)
print('mma',mma_data_transposed)


# Define the epochs
epochs = list(range(5))
# Plot Training Accuracy
plt.figure(figsize=(10, 6))
plt.plot(epochs, xyz_data_transposed[0], label='XYZ Features Train Loss')
plt.plot(epochs, hks_data_transposed[0], label='HKS Features Train Loss')
plt.plot(epochs, shot_data_transposed[0], label='SHOT Features Train Loss')
plt.plot(epochs, lps_data_transposed[0], label='LPS Features Train Loss')
plt.plot(epochs, mma_data_transposed[0], label='Combined Features (SHOT,LPS,WKS,SGWS,XYZ) Train Loss')


plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.title('Training Loss Comparison')
plt.legend()
plt.grid(True)
plt.xticks(range(5))  # To show epochs from 0 to 4 in ints
plt.show()

# Plot Test Accuracy
plt.figure(figsize=(10, 6))
plt.plot(epochs, xyz_data_transposed[1], label='XYZ Features Test Loss')
plt.plot(epochs, hks_data_transposed[1], label='HKS Features Test Loss')
plt.plot(epochs, shot_data_transposed[1], label='SHOT Features Test Loss')
plt.plot(epochs, lps_data_transposed[1], label='LPS Features Test Loss')
plt.plot(epochs, mma_data_transposed[1], label='Combined Features (SHOT,LPS,WKS,SGWS,XYZ) Test Loss')

plt.xlabel('Epochs')
plt.ylabel('Test Loss')
plt.title('Test Loss Comparison')
plt.legend()
plt.grid(True)
plt.xticks(range(5))
plt.show()

# Plot Geodesic Error
plt.figure(figsize=(10, 6))
plt.plot(epochs, xyz_data_transposed[2], label='XYZ Features Geodesic Error')
plt.plot(epochs, hks_data_transposed[2], label='HKS Features Geodesic Error')
plt.plot(epochs, shot_data_transposed[2], label='SHOT Features Geodesic Error')
plt.plot(epochs, lps_data_transposed[2], label='LPS Features Geodesic Error')
plt.plot(epochs, mma_data_transposed[2], label='Combined Features (SHOT,LPS,WKS,SGWS,XYZ) Geodesic Error')

plt.xlabel('Epochs')
plt.ylabel('Geodesic Error')
plt.title('Geodesic Error Comparison')
plt.legend()
plt.grid(True)
plt.xticks(range(5))
plt.show()
```
