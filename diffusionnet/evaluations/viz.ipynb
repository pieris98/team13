```python
import matplotlib.pyplot as plt

# Assuming your text data is in 'xyz_features.txt' and 'hks_features.txt'

def read_data(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()[1:]  # Skip the header line
        data = [list(map(float, line.strip().split(';'))) for line in lines]
    return data

# Read the data from files
xyz_data = read_data('diffnet_xyz.txt')
hks_data = read_data('diffnet_hks.txt')

# Convert the data to a format suitable for plotting
# Transposing the list of lists to get columns of data
xyz_data_transposed = list(map(list, zip(*xyz_data)))
hks_data_transposed = list(map(list, zip(*hks_data)))

# Define the epochs
epochs = list(range(5))

# Plot Training Accuracy
plt.figure(figsize=(10, 6))
plt.plot(epochs, xyz_data_transposed[0], label='XYZ Features Train Accuracy')
plt.plot(epochs, hks_data_transposed[0], label='HKS Features Train Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Training Accuracy')
plt.title('Training Accuracy Comparison')
plt.legend()
plt.grid(True)
plt.xticks(range(5))  # To show epochs from 0 to 4 in ints
plt.show()

# Plot Test Accuracy
plt.figure(figsize=(10, 6))
plt.plot(epochs, xyz_data_transposed[1], label='XYZ Features Test Accuracy')
plt.plot(epochs, hks_data_transposed[1], label='HKS Features Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy Comparison')
plt.legend()
plt.grid(True)
plt.xticks(range(5))
plt.show()

# Plot Geodesic Error
plt.figure(figsize=(10, 6))
plt.plot(epochs, xyz_data_transposed[2], label='XYZ Features Geodesic Error')
plt.plot(epochs, hks_data_transposed[2], label='HKS Features Geodesic Error')
plt.xlabel('Epochs')
plt.ylabel('Geodesic Error')
plt.title('Geodesic Error Comparison')
plt.legend()
plt.grid(True)
plt.xticks(range(5))
plt.show()
```
